```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/solaojp/Venezuela-crisis/master/data.csv")
```

```python
data.head()
```

```python
df = data.drop(data.columns[[0]], axis=1)  
df = df.fillna(0)
```

```python
df = df.replace(to_replace='democracy', value='Democracy', regex=True)
```

```python
plt.figure(figsize=(20,12))
x = sns.countplot(x='Political model', data=df)
plt.xlabel('Political model')
plt.ylabel('Count')
x.set_xticklabels(x.get_xticklabels(), rotation=90)
plt.show()
```

```python
# The following is a list of parties which are democratic socialist or partly democratic socialist currently having representation in the legislature of their country.
df1 = df[df['Political model'].str.contains("Democratic")==True]
df1
```

```python
df2 = df[df['Unrest Cases']!= 0]
df2
```

```python
plt.figure(figsize=(20,12))
x1 = sns.barplot(x='Country Name',y = 'Unrest Cases',data=df2)
plt.xlabel('Unrest Cases against goverment')
plt.ylabel('Unrest Cases')
x1.set_xticklabels(x1.get_xticklabels(), rotation=90)
plt.show()
```
